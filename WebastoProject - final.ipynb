# Between 2026 - 2028 - Current time

Filtering out the records for the current time based on the GPD data

```python
# Creating the name of the columns which we want to focus on
Volume_cols = [f'Volume {year}' for year in range(2026, 2029)]

# Filtering out the records that are for the current time
Current_GPD_df = GPD_df.copy()

# Summing up the total volume in the current time
Current_GPD_df['Total volume - current time'] = Current_GPD_df[Volume_cols].sum(axis = 1)

# Filtering out the records for the current time
Current_GPD_df = Current_GPD_df[Current_GPD_df['Total volume - current time'] > 0].reset_index(drop = True)

Current_GPD_df.head()
```

Filtering out the data in the current time based on the IHS data.

```python
# Creating the name of the columns which we want to focus on
CY_cols = [f'CY {year}' for year in range(2026 , 2029)]

# Filtering out the records that are for the current time
Current_IHS_df = IHS_df.copy()

# Summing up the total volume in the current time
Current_IHS_df['Total volume global - current time'] = Current_IHS_df[CY_cols].sum(axis = 1)

# Filtering out the records for the current time
Current_IHS_df = Current_IHS_df[Current_IHS_df['Total volume global - current time'] > 0].reset_index(drop = True)

Current_IHS_df.head()
```

### Top 5 platforms

Here we are going to plot our 4 different dataframes to show which platforms had the highets total volume for each __Acquisition status__

```python
# Filtering out the records that we have the platform
Current_GPD_platform_df = Current_GPD_df[~Current_GPD_df['Platform'].isnull()][['Acquisition Status', 'Platform', 'Total volume - current time']]

# Compute the total volume for each platform for each Acquisition status
Current_GPD_platform_df = Current_GPD_platform_df.groupby(by = ['Acquisition Status', 'Platform']).sum().reset_index().sort_values('Total volume - current time', ascending = False)

# Get the different values for status of the acquisition
Status_values = list(Current_GPD_platform_df['Acquisition Status'].unique())

for status in Status_values:
  print('Acquisition status: ', status)
  # filter the records for that specific status
  filtered_df = Current_GPD_platform_df[Current_GPD_platform_df['Acquisition Status'] == status]

  # display the top 5 platforms for each acquistion status
  display(filtered_df.sort_values('Total volume - current time', ascending = False).head().reset_index(drop = True))
  print('\n', '-' * 30)
```

Here we will see the top 5 platforms in the global market.

```python
# Summing up the total volume for all of the platforms in between 2026 and 2028
Current_IHS_platforms_df = Current_IHS_df[['VP: Platform','Total volume global - current time']].groupby(by = 'VP: Platform').sum().sort_values('Total volume global - current time', ascending = False).reset_index()

# Identifying the top 5 platforms in the matter of the total global volume in between 2026 and 2028
Current_IHS_platforms_df.head()
```

#### All acquisition status

The plot showing the top 5 by considering the total volume between 2026 and 2028 based on the __GPD data__

```python
# The name of the platforms that had the highet total volume
Top_platform_names = list(Current_GPD_platform_df.groupby(by = ['Platform']).sum().reset_index().sort_values('Total volume - current time', ascending = False)['Platform'])[:5]

# Filtering out the reocords only for those top platfomrs
Filtered_platforms = Current_GPD_platform_df[Current_GPD_platform_df['Platform'].isin(Top_platform_names)].copy()

# Sorting the records based on the total volume
Filtered_platforms.sort_values('Total volume - current time', ascending = False, inplace = True)

# Plotting the distribution of those platforms based on the acquisition status
fig = px.bar(Filtered_platforms, x='Platform', y='Total volume - current time',
             color='Acquisition Status', barmode='group',
             height=400, title = 'Top 5 platforms with the highest total volume - (2026 - 2028)',  text_auto='.2s',
            color_discrete_map={ 'Competition Booked': 'red', 'Competition Unbooked': '#9725cc',
                                 'Webasto Booked': 'blue', 'Webasto Unbooked' : '#10b560'})
fig.show()
```
